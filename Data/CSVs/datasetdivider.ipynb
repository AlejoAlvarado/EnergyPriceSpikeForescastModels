```python
import pandas as pd
import os

df = pd.read_csv("aeso_merged_2020_2025.csv")
print("Columns loaded correctly:", df.columns[:5].tolist())
print("Shape:", df.shape)

n = 4
chunk_size = len(df) // n

for i in range(n):
    start = i * chunk_size
    end = len(df) if i == n - 1 else (i + 1) * chunk_size
    part = df.iloc[start:end]
    filename = f"aeso_part_{i+1}.csv"
    part.to_csv(filename, index=False)
    size_mb = os.path.getsize(filename) / 1e6
    print(f"Part {i+1}: {len(part)} rows, {size_mb:.1f} MB")
```
